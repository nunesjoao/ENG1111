# **Aula 2 - Introdução a PANDAS**

Pandas é uma biblioteca de código fonte aberto feita para trabalhar com conjuntos de dados complexos. Suas funções podem ser utilizadas para importar, filtrar e extrair informações de um conjunto de dados. Nesta aula veremos os seguintes tópicos:




* Series
*  Dataframe
*  Importando dados CSV
*  Manipulando um dataframe
* Convertendo um dataframe para matriz



```python

# referência biblioteca https://pandas.pydata.org/docs/search.html?q=display
# Importando a biblioteca pandas

import pandas as pd

# as pd cria uma apelido para pandas, tornando a sintaxe mais prática
```

# Series

Series é uma estrutura de Pandas que se assemelha a uma coluna em uma tabela,
é um array de uma dimensão que pode armazenar dados de qualquer tipo.

Exemplo:



```python
#declarando uma lista 

aluno = ['joao','pedro','zeca']

#converte lista para formato Series
coluna = pd.Series(aluno)

print(coluna)
```

```python
#acessando os valores por meio dos seus índices

print(coluna[1])
```

```python
#Criando Labels

# Se nada for especificado, os índices serão criados automaticamente de forma sequencial. Podemos criar os próprios índices

coluna2 = pd.Series(aluno, index = ["J","P","Z"])
                  
print(coluna2)
```

```python
# Acessando os elementos por meio dos índices

print(coluna2["J"])
```

```python
# Criando um Series a partir de um dicionário

temperos = {"coentro":200,"salsa":130,"manjerona":120}
 
tempdata = pd.Series(temperos);

print(tempdata['coentro'])
```

# Dataframes

Dataframes é uma estrutura que permite armazenar uma tabela multidimensional.



```python
# declarando um dicionário

carros = {
    "carros":['fiat','chevrolet','audi'],
    "preços":[30.000, 45.000,100.000]
}

#convertendo para Dataframe

dados = pd.DataFrame(carros)

print(dados)
```

```python
# retornando uma linha

print(dados.loc[0])

#retornando mais de uma linha

print(dados.loc[[0,2]])
```

## Importando arquivos CSV 

Pandas pode importar diretamente arquivos CSV e JSON. CSV (Comma separated values) é um tipo de arquivo txt que armazena dados separados por vírgulas. Embora não seja padronizado como JSON, o formato é utilizado por diversos programas para exportar dados.

```python
# Vamos importar um arquivo CSV diretameente da Internet utilizando Pandas

import os

pd.set_option('display.max_columns',7)
data = pd.read_csv("http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/reportes_especies_2011_2021.csv")

# display apresenta os dados de forma mais estruturada.

display(data[0:5])




```

```python
# obtendo informações a respeito do dataset

print(data.info())

```

# Manipulando um dataframe

Frequentemente os dados devedrão ser tratados antes que possamos utiliza-los na nossa aplicação. Pandas dispões de diversas ferramentas para essa finalidade.

### Limpando células vazias

```python
#limpa a célula com

nova_data = data.dropna()

display(nova_data)

# mostrando as informações do novo dataset
print(nova_data.info())
```

## Substituindo valores vazios por valores pré determinados nas colunas especificadas

```python

# inplace grava os novos valores no dataset

data["amostra"].fillna(0,inplace = True)

display(data)
```

## Substituindo valores vazios pela média

```python


#lendo arquivo csv do google drive


carro = pd.read_csv("/content/drive/MyDrive/auto-mpg.csv")

#Calcula a média de todos os campos tamanho.x

media = carro["displacement"].mean()
print(media)

#atualiza os campos vazios com a média

carro["displacement"].fillna(media, inplace = True)

display(carro)



```

## Descartando valores

```python
carro2 = pd.read_csv("/content/drive/MyDrive/auto-mpg.csv")

# descarta a coluna cylinders e atualiza o dataframe com inplace

carro2.drop(columns=['origin','horsepower'], inplace = True)

display(carro2)
```

## Concatenando colunas ou linhas

```python
carro3 = pd.read_csv("/content/drive/MyDrive/auto-mpg.csv")

#seleciona coluna cylinder
col_cylinder=carro3['cylinders']

#seleciona coluna weight
col_weight = carro3['weight']

#cria novo dataframe concatenando as colunas cylinder e weight
resultado = pd.concat([col_cylinder,col_weight],axis=1)

#configurando opções para o display

pd.set_option('display.max_columns', 0)
pd.set_option('display.max_rows', 5)


display(resultado)
```
